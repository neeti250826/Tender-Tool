```python
import pandas as pd

contracts = pd.read_csv("output_260126/step3_contracts_Sheet2.csv")
items = pd.read_csv("output_260126/step3_dialog_items_Sheet2.csv")


# Aggregate dialog items per contract
agg = (items
       .groupby(["URL", "contract_index", "Número de contrato"], as_index=False)
       .agg(
           items_count=("Subtotal (num)", "size"),
           subtotal_sum=("Subtotal (num)", "sum"),
           iva_sum=("IVA (num)", "sum"),
           otros_sum=("Otros impuestos (num)", "sum"),
           total_sum=("Total (num)", "sum"),
       ))

# Join against contract totals
chk = contracts.merge(
    agg,
    on=["URL", "contract_index", "Número de contrato"],
    how="left"
)

chk["diff_subtotal_vs_contract"] = chk["subtotal_sum"] - chk["Importe total sin impuestos (num)"]
chk["diff_total_vs_subtotal_plus_tax"] = chk["total_sum"] - (chk["subtotal_sum"] + chk["iva_sum"] + chk["otros_sum"])

print(chk[[
    "contract_index", "Número de contrato", "Licitante",
    "dialog_items_count", "items_count",
    "Importe total sin impuestos (num)", "subtotal_sum", "diff_subtotal_vs_contract",
    "iva_sum", "otros_sum", "total_sum", "diff_total_vs_subtotal_plus_tax"
]])
```
