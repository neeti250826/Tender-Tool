This notebook runs the GeBIZ scraper with **advanced search filters**.

Tip: Use `ACTIVE_ONLY` / `INACTIVE_ONLY` to quickly switch between open vs closed/awarded opportunities.

```python
# @title Filters
DATE_FROM = '2026-03-01'  # @param {type:'string'}
DATE_TO = '2026-03-14'  # @param {type:'string'}
QUERY = ''  # @param {type:'string'}
OUTPUT_TARGET = './_local_outputs'  # @param {type:'string'}
PROJECT_NAME = 'MDT_2026'  # @param {type:'string'}
WEBSITE_ID = 'SG_GEBIZ'  # @param {type:'string'}
SOURCE_LABEL = 'Singapore GeBIZ'  # @param {type:'string'}
REGION = 'EMEA'  # @param ['EMEA','LATAM']

USE_ADVANCED_SEARCH = True  # @param {type:'boolean'}
ACTIVE_ONLY = True  # @param {type:'boolean'}
INACTIVE_ONLY = False  # @param {type:'boolean'}

# Optional advanced keyword helpers (used when USE_ADVANCED_SEARCH=True)
TITLE_ALL = ''  # @param {type:'string'}
TITLE_ANY = ''  # @param {type:'string'}
TITLE_NONE = ''  # @param {type:'string'}
TITLE_EXACT = ''  # @param {type:'string'}
DOCUMENT_NO = ''  # @param {type:'string'}
REFERENCE_NO = ''  # @param {type:'string'}
```

```python
# @title Advanced filters help
!python "gebiz_scraper.py" --list-advanced-filters
```

```python
# Run
adv = []
if USE_ADVANCED_SEARCH: adv.append('--use-advanced-search')
if ACTIVE_ONLY: adv.append('--active-only')
if INACTIVE_ONLY: adv.append('--inactive-only')
if TITLE_ALL: adv += ['--title-all', TITLE_ALL]
if TITLE_ANY: adv += ['--title-any', TITLE_ANY]
if TITLE_NONE: adv += ['--title-none', TITLE_NONE]
if TITLE_EXACT: adv += ['--title-exact', TITLE_EXACT]
if DOCUMENT_NO: adv += ['--document-no', DOCUMENT_NO]
if REFERENCE_NO: adv += ['--reference-no', REFERENCE_NO]
cmd = [
    'python', 'gebiz_scraper.py',
    '--date-from', DATE_FROM,
    '--date-to', DATE_TO,
    '--query', QUERY,
    '--output-target', OUTPUT_TARGET,
    '--project-name', PROJECT_NAME,
    '--website-id', WEBSITE_ID,
    '--source-label', SOURCE_LABEL,
    '--region', REGION,
] + adv
import subprocess
print(' '.join([str(x) for x in cmd]))
subprocess.run(cmd, check=False)
```
